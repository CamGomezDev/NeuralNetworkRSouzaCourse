### To teach machine learning such task one needs: Recurrent Neural Network not Feedforward-Neural-Network (FNN)

### FNN

* In a feed-forward neural network that we have seen so far (regression problem or classification problem), the information only moves in one direction — from the input layer, through the hidden layers, to the output layer. The information moves straight through the network and never touches a node twice!!


* Feed-forward neural networks have no memory of the input they receive and are bad at predicting what’s coming next. Because a feed-forward network only considers the current input, it has no notion of order in time. It simply can’t remember anything about what happened in the past except its training.


* Imagine you have a normal feed-forward neural network and give it the word "Artificial" as an input and it processes the word character by character. By the time it reaches the character "i," it has already forgotten about "A," "r" and "t," which makes it almost impossible for this type of neural network to predict which character would come next.

### RNN

* A recurrent neural network, however, is able to remember those characters because of its internal memory. It produces output! Then it copies that output and loops it back into the network.

***Simply put: recurrent neural networks add the immediate past to the present.***

* Therefore, a RNN has two inputs: the present and the recent past. This is important because the sequence of data contains crucial information about what is coming next, which is why a RNN can do things other algorithms can’t.


* Like many other deep learning algorithms, recurrent neural networks are relatively old. They were initially created in the 1980’s, but only in recent years have we seen their true potential. An increase in computational power along with the the massive amounts of data that we now have to work with, and the invention of long short-term memory (LSTM) in the 1990s, has really brought RNNs to the foreground.


* "Recurrent neural networks (RNN) are the state of the art algorithm for sequential data and are used by Apple's Siri and and Google's voice search and Google word prediction mechanism. It is the first algorithm that remembers its input, due to an internal memory, which makes it perfectly suited for machine learning problems that involve sequential data. It is one of the algorithms behind the scenes of the amazing achievements seen in deep learning over the past few years. In this post, we'll cover the basic concepts of how recurrent neural networks work, what the biggest issues are and how to solve them."

## Difference in architecture between FNN and RNN

The main difference in RNN and Forward NN is that in each neuron of RNN, the output of previous time step is feeded as input of the next time step. This makes RNN be aware of time (at least time units) while the Feedforward has none. For example, in the handwritten digits classification, you have the input and output. There is no difference between time 1st or time 100th because the network has the same input and output. RNN is usually used to describe a sequence (can be time sequence or a context). For example, reading a big text file and output the next character. As you read more, the knowledge you accumulate through your previous timestep makes the network aware of the context. So RNN will have better performance than vanilla Feedforward because Feedforward has no better idea what to output, given all the previous outputs.

## Rolled y Unrolled RNN
The image below illustrates an unrolled RNN. On the left, the RNN is unrolled after the equal sign. Note there is no cycle after the equal sign since the different time steps are visualized and information is passed from one time step to the next. This illustration also shows why a RNN can be seen as a sequence of neural networks.

<img src="imgs/unrolled-rnn_0_2.png" width="600" />

RNN has input, hidden state and output. Hidden state receives two output, from the input and the previous output.

* Input: x(t) is taken as the input to the network at time step t. For example, x1,could be a one-hot vector corresponding to a word of a sentence.

* Hidden state: h(t) represents a hidden state at time t and acts as “memory” of the network. h(t) is calculated based on the current input and the previous time step’s hidden state: h(t) = f(U x(t) + W h(t−1)). The function f is taken to be a non-linear transformation such as tanh, ReLU.

* Weights: The RNN has input to hidden connections parameterized by a weight matrix U, hidden-to-hidden recurrent connections parameterized by a weight matrix W, and hidden-to-output connections parameterized by a weight matrix V and all these weights (U,V,W) are shared across time.

* Output: o(t) illustrates the output of the network. 

RNNs are feedback neural networks, which means that the links between the layers allow for feedback to travel in a reverse direction. This helps the neural network to learn contextual information. **For repeated patterns, more weight is applied to the previous patterns than the one being currently evaluated**. 

## Pros and cons of RNN

* Advantage of an RNN

  *    Input of any length can be processed.
  *    Model size does not increase if the input is larger.
  *    RNN model is trained to remember each information throughout the time which is very helpful in any time series predictor.

* Disadvantages of an RNN

  *    Computation is slow, due to its recurrent nature.
  *    Model training can be difficult
  *    Not suited for predicting long horizons.

## Problems with RNN:

* EXPLODING GRADIENTS
  * Exploding gradients are when the algorithm, without much reason, assigns a stupidly high importance to the weights. Fortunately, this problem can be easily solved by truncating or squashing the gradients or regularization
  

* VANISHING GRADIENTS - more serious
  * Vanishing gradients occur when the values of a gradient are too small and the model stops learning or takes way too long as a result. This was a major problem in the 1990s and much harder to solve than the exploding gradients. Fortunately, it was solved through the concept of LSTM by Sepp Hochreiter and Juergen Schmidhuber.

## Long Short-Term Memory (LSTM)

* Long short-term memory networks (LSTMs) are an extension for recurrent neural networks, which basically extends the memory. Therefore it is well suited to learn from important experiences that have very long time lags in between.


* LSTMs enable RNNs to remember inputs over a long period of time. This is because LSTMs contain information in a memory, much like the memory of a computer. The LSTM can read, write and delete information from its memory.


* In an LSTM you have three gates: input, forget and output gate. These gates determine whether or not to let new input in (input gate), delete the information because it isn’t important (forget gate), or let it impact the output at the current timestep (output gate). Below is an illustration of a RNN with its three gates (basically like human brain...):

<img src="imgs/rnn-three-gates.png" width="600" />

The gates in an LSTM are analog in the form of sigmoids, meaning they range from zero to one. The fact that they are analog enables them to do backpropagation.

The problematic issues of vanishing gradients is solved through LSTM because it keeps the gradients steep enough, which keeps the training relatively short and the accuracy high.

# Recurrent Neural Network and Time Series

    
*     Time series is a sequence of observations taken sequentially in time.
*     Forecast means making predictions about a future event.

Time series forecast has been in use across multiple industries for quite some time; it is commonly used in every industry to guide future decision, for example in retail sales forecast is very important, so that the raw material can be procured accordingly. The most famous example is weather forecasting, where based on the pattern in the past and recent changes, the future can be predicted. These predictions are very important and usually are the first step to solve other problem, as plan the power generations to avoid unnecessary power disruptions or overproduction.

# Popular applications of time series forecasting
* Stock prices forecasting - Many advanced Time Series Forecasting models are used to predict stock prices, since in the historical sequences there is a lot of noise and a high uncertainty in the information, that may depend on several factors not always closely related to the stock market.

* Weather Prediction - Time Series Forecasting models are widely used to
improve the accuracy of weather forecasts.

* Forecasting Traffic - Travel planning applications use Time Series Forecasting models to predict traffic on the roads, in order to decide more accurately the fastest way to arrive to the selected destination.

* Churn - Many companies use Time Series Forecasting models to predict in which month they can expect a higher turnover, so that they can implement the best employee retention plans.

* Text generation - When we write a text on the smartphone, it looks at the last written words and at the typed letters, and suggests the next letters or even the whole word.

# RNN practical structure
Typically it would be:

* batch size, 
* the number of steps 
* number of features. 

So the data has to be 3D. `[(samples, time_steps, features)]`

The number of steps depicts the number of time steps/segments you will be feeding in one line of input of a batch of data that will be fed into the RNN. 
RNNs is the ability to model short term dependencies.

````python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt
from matplotlib import rc


RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
````

````python
time = np.arange(0, 100, 0.1)
sin = np.sin(time) + np.random.normal(scale=0.5, size=len(time))
````

````python
sin.shape
````

````python
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.plot(time, sin, label='sine (with noise)');
ax.legend();
ax.grid(True)
````

````python
df = pd.DataFrame(dict(sine=sin), index=time, columns=['sine'])
````

````python
df.tail()
````

````python
train_size = int(len(df) * 0.8)
test_size = len(df) - train_size
train, test = df.iloc[0:train_size], df.iloc[train_size:len(df)]
print(len(train), len(test))
````

````python
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.plot(train['sine'])
ax.plot(test['sine'],color='y')
ax.grid(True)
````

# Intuitively:

#### Time series should be divided into overlapping time windows like that:
```
Data Set  : [1,2,3,4,5,6] => 
X_train is: [[1,2,3],[2,3,4],[3,4,5],[4,5,6]]
y_train is: [    [4],    [5],    [6],    [7]]    
```

````python
# some generic function
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
````

````python
time_steps = 10

# reshape to [samples, time_steps, n_features]

X_train, y_train = create_dataset(train, train['sine'], time_steps)
X_test, y_test = create_dataset(test, test['sine'], time_steps)

print(X_train.shape, y_train.shape)
````

# Lets build simple RNN network

````python
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
````

````python
model = keras.Sequential([SimpleRNN(units=10, input_shape=(X_train.shape[1], X_train.shape[2])),
                                    Dense(units=1)])
model.compile(loss='mean_squared_error',optimizer=keras.optimizers.Adam(0.001))
````

````python
model = keras.Sequential([LSTM(units=128, input_shape=(X_train.shape[1], X_train.shape[2])),
                          Dense(units=1)])
model.compile(loss='mean_squared_error',optimizer=keras.optimizers.Adam(0.001))
````

````python
history = model.fit(
                     X_train, y_train,
                     epochs=20,
                     batch_size=16,
                     validation_split=0.1,
                     verbose=1,
                     shuffle=False
                     )
````

````python
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.plot(history.history['loss'])
ax.plot(history.history['val_loss'])
ax.grid(True)
````

````python
X_test.shape
````

````python
fig,ax=plt.subplots(1,1)
ax.plot(np.arange(0,len(X_test[0])),X_test[0],c='b',marker='o',ms=10)
ax.grid(True)
````

So `X_test[0]`, is the sequences of 10 points and we ask what is the 11th

`X_test` in total has 190 sequences are like this. So there will 190 predictions

````python
y_pred = model.predict(X_test)
````

#### notice the shape of the test data (also 3D)

````python
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.plot(np.arange(0, len(y_train)), y_train, 'b', label="train")
ax.plot(np.arange(len(y_train), len(y_train) + len(y_test)), y_test, c='y',marker='.', label="true")
ax.plot(np.arange(len(y_train), len(y_train) + len(y_test)), y_pred, 'k', label="prediction")
ax.set_ylabel('Value')
ax.set_xlabel('Time Step')
ax.legend()
ax.grid(True)
````

````python
fig,ax=plt.subplots(1,1,figsize=(10,8))
ax.plot(y_test, color='y', marker='.', label="true")
ax.plot(y_pred, 'k', label="prediction")
ax.set_ylabel('Value')
ax.set_xlabel('Time Step')
ax.legend()
ax.grid(True)
````
