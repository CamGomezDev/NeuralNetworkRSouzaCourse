```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt 
  
## generate the data and plot it for an ideal normal curve
  
## x-axis for the plot
x_data = np.arange(0, 10, 0.001)
  
## y-axis as the gaussian
y_data_1 = stats.norm.pdf(x_data, 4, 0.3)
y_data_2 = stats.norm.pdf(x_data, 6, 0.3)

## plot data
plt.plot(x_data, y_data_1)
plt.plot(x_data, y_data_2)
```

```python
import numpy as np
import scipy as sp
from scipy import stats
import matplotlib.pyplot as plt 
  
## generate the data and plot it for an ideal normal curve
  
## x-axis for the plot
x_data = np.arange(0, 10, 0.001)
  
## y-axis as the gaussian
y_data_1 = stats.norm.pdf(x_data, 4, 2.3)
y_data_2 = stats.norm.pdf(x_data, 6, 2.3)

## plot data
plt.plot(x_data, y_data_1)
plt.plot(x_data, y_data_2)
```
