# Exercise 1
* the file 'mnist_test.csv' contains 10000 images of digits in grayscale saved as a rows in the csv file
* first column contains value of the digits then 784 columns correspond to pixels
* make an array of plots of 25 randomly chosen images (5x5)
* each plot has to have a title corresponding to a digit

```python
data = pd.read_csv("mnist_test.csv")
data
```

```python
data.shape # exploration of the data
```

### Steps to do it (explanation):
* convert type(DataFrame) to type(numpy.ndarray). 
* define the labels by taking the column values in the column called "label" and convert it in a numpy array.
* every row in the dataset is divided by 28 parts that represent every pixel in the image using np.reshape, it has to start in 1 to not take the label number.
* pick a random integer number from 0 to 9999 (lenght of the DataFrame).
* use this random integer number to take a random label and using matplotlib, draw the digit.

```python
data_np = data.to_numpy() # data converted to np array
label = data["label"].to_numpy() # labels converted to np array
```

```python
fig, ax = plt.subplots(5,5, figsize=(10,10),constrained_layout=True) # figure 5x5 images
for i in range(5): # horizontal axis
    for j in range(5): # vertical axis
        r = np.random.randint(len(data)) # pick a random number 
        image = np.reshape(data_np[r][1:], [28,28])
        ax[i,j].imshow(image,cmap="Greys") # divide the random row in 28x28 arrays and draw in a plot. 
        ax[i,j].set_title( "Number: "+str(label[r]) ) # Title of the respective number
```
