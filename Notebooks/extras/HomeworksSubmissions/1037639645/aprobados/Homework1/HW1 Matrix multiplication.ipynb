```python
#Libraries

import numpy as np
```

```python
#Matrix definition

#Matrix 3x2
A=np.array([[4,5],[2,3],[4,7]])

#Matrix 2x3
B=np.array([[3,5,6],[4,3,1]])
```

```python
print(A)
print(B)
```

```python
#Multtiplication using loops:

#Ending matrix

R=np.zeros((3,3))

#Iterating through rows of A

for i in range(len(A)):
    #Iterating through columns of B
    for j in range(len(B[0])):
        #Iterating through rows of B
        for k in range(len(B)):
            R[i][j]+=A[i][k]*B[k][j]
            
```

```python
print(R)
```

```python
R=A@B
print(R)
```
