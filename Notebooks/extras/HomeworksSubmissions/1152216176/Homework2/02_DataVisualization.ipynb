## Exercise 1

* the file 'mnist_test.csv' contains 10000 images of digits in grayscale saved as a rows in the csv file
* first column contains value of the digits then 784 columns correspond to pixels
* make an array of plots of 25 randomly chosen images (5x5)
* each plot has to have a title corresponding to a digit

# Load python libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the data
images_df = pd.read_csv('mnist_test.csv')

images_df.shape
```

```python
#transform to numpy array
all_images = images_df.to_numpy()

#generate a random 25 length array to select the random numbers
random_selection = np.random.randint(0,images_df.shape[0],25)
selected_images = all_images[random_selection]
```

```python
#making the plot array
fig, ax = plt.subplots(5, 5, figsize=(15, 15))
#this is just so that the titles and graphics do not overlap
plt.subplots_adjust(hspace=0.4)

for i in range(5):
    for j in range(5):
        #index goes from 0 to 24
        index = 5*i+j
        label = selected_images[index][0]
        #reshape arrays
        image = selected_images[index][1:].reshape(28, 28)
        ax[i,j].set_title("label: " + str(label))
        ax[i,j].imshow(image,cmap='gray')
```

## Exercise 2

* Sklearn digits dataset (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html) contains 1797 images of digits in grayscale
* download the dataset (command is below) and:
* make an array of plots of 25 randomly chosen images (5 images x 5 images)
* each plot has to have a title corresponding to a digit

```python
from sklearn import datasets
digits = datasets.load_digits() # digits is a dictionary
```

```python
#save images an labels in independent arrays
digit_images = digits['data']
digit_labels = digits['target']

digit_images.shape
```

```python
#generate a random 25 length array to select the random numbers
random_selection = np.random.randint(0,digit_images.shape[0],25)
selected_images = digit_images[random_selection]
selected_labels = digit_labels[random_selection]
```

```python
#making the plot array
fig, ax = plt.subplots(5, 5, figsize=(15, 15))
#this is just so that the titles and graphics do not overlap
plt.subplots_adjust(hspace=0.4)

for i in range(5):
    for j in range(5):
        #index goes from 0 to 24
        index = 5*i+j
        label = selected_labels[index]
        #reshape arrays
        image = selected_images[index].reshape(8, 8)
        ax[i,j].set_title("label: " + str(label))
        ax[i,j].imshow(image,cmap='gray')
```

The CIFAR-10 dataset consists of, in total, 60000 color (three channels) images, each with one of 10 labels: airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck. For an introduction and a download.

* make an array of plots of 25 randomly chosen images (5 images x 5 images)
* each plot has to have a title with a corresponding label
* convert the color images to grayscale by averaging one of the dimensions and again display the images
* make an array of plots of 25 randomly chosen images (5 images x 5 images)
* each plot has to have a title with a corresponding label

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
```

```python
X_data.shape
```

```python
#generate a random 25 length array to select the random numbers
random_selection = np.random.randint(0,X_data.shape[0],25)
selected_images = X_data[random_selection]
selected_labels = y_data[random_selection]

image_label = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]
```

```python
#making the plot array
fig, ax = plt.subplots(5, 5, figsize=(15, 15))
#this is just so that the titles and graphics do not overlap
plt.subplots_adjust(hspace=0.4)

for i in range(5):
    for j in range(5):
        #index goes from 0 to 24
        index = 5*i+j
        label = image_label[int(selected_labels[index])]
        image = selected_images[index]
        ax[i,j].set_title("label: " + label)
        ax[i,j].imshow(image)
```

```python
X_data_gray = np.average(X_data,-1)
X_data_gray.shape
```

```python
random_selection = np.random.randint(0,X_data_gray.shape[0],25)
selected_images_gray = X_data_gray[random_selection]
selected_labels = y_data[random_selection]
```

```python
#making the plot array
fig, ax = plt.subplots(5, 5, figsize=(15, 15))
#this is just so that the titles and graphics do not overlap
plt.subplots_adjust(hspace=0.4)

for i in range(5):
    for j in range(5):
        #index goes from 0 to 24
        index = 5*i+j
        label = image_label[int(selected_labels[index])]
        image = selected_images_gray[index]
        ax[i,j].set_title("label: " + label)
        ax[i,j].imshow(image,cmap='gray')
```
