Use scikit-learn to perform digit classification using data from Homework 2 (mnist_test.csv)

```python
#Load python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
#reading the data
images_df = pd.read_csv('mnist_test.csv')

images_df
```

```python
#separate the data into X_data (images) and y_data (label), transform them into numpy arrays
X_data = images_df.drop('label',axis=1).to_numpy()
y_data = images_df['label'].to_numpy()
```

```python
#split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=1046)
```

```python
#use the data from the training set with sklearn LogisticRegression to train the number classifier
digit_classifier = LogisticRegression()
digit_classifier.fit(X_train, y_train)
```

```python
#we get the predictions of the model on the test data set (y_predict) and compare them with the real labels (y_test)
#to obtain the score for each digit

y_predict = digit_classifier.predict(X_test)
print(classification_report(y_test, y_predict))
```

We can see that the digits that present a lower score are 5 and 8, both for precision and for recall, in this case it could be due to the similarity in the shape of both digits that the model tends more to confuse them.
