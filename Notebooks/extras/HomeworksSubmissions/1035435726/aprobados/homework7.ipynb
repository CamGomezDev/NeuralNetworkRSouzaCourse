Let's do a manual clean of the outliers. We will choose the column bedroom

```python
cols = ["bedrooms", "sqft_lot", "sqft_living", "sqft_basement"]
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3-Q1
print(Q3+1.5*IQR)
```

```python
min_vals = Q1 - 1.5 * IQR
max_vals = Q1 + 1.5 * IQR
```

```python
mask = ((df.bedrooms < min_vals["bedrooms"]) | (df.bedrooms> max_vals["bedrooms"]))
```

```python
df_New = df[~mask]
df_New[mask]
```

```python
df_New[["bedrooms"]].plot(kind='box', figsize=(5,5),title='Boxplots')
```

## NN Predictions

```python
X_data = df_New.drop(["id","date","price","price_log", "date_time"], axis=1)
y_data = df_New["price_log"]
```

```python
X_data
```

```python
from sklearn.model_selection import train_test_split
(X_train,X_test,y_train,y_test)=train_test_split(X_data,y_data,test_size=0.1,random_state=1234)
```

```python
import tensorflow as tf
tf.random.set_seed(1234)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from tensorflow.keras import regularizers
```

```python
def get_model(input_shape):
    model=Sequential([InputLayer(input_shape=input_shape),
                Dense(units=128,activation='relu'),
                Dense(units=128,activation='relu'),
                Dense(units=128,activation='relu'),
                Dense(units=1,activation='linear')
    ])
    return model
```

```python
def compile_model(model):
    model.compile(optimizer='adam',loss='mse',metrics=['mae','mape'])
```

```python
X_train.shape
```

```python
model=get_model(input_shape=X_train.shape[1])
model.summary()
compile_model(model)
```

```python
history=model.fit(X_train,y_train,batch_size=64, epochs=100, validation_split=0.15,verbose=1)
scores=model.evaluate(X_test,y_test)
print(scores)
```

```python
# Plot the training and validation loss
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss vs. epochs')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()
```

```python
# Plot the training and validation loss
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(history.history['mape'])
plt.plot(history.history['val_mape'])
plt.title('Mape vs. epochs')
plt.ylabel('Mape')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()
```

The results have improved. Now let us try with using regularization. l1 and l2 and dropout

```python
def get_regularized_model(input_shape,wd,drate):
    model=Sequential([InputLayer(input_shape=input_shape),
                Dense(128, kernel_regularizer = regularizers.l1(wd), activation="relu"),
                Dropout(drate),
                Dense(128, kernel_regularizer = regularizers.l1(wd), activation="relu"),
                Dropout(drate),
                Dense(128, kernel_regularizer = regularizers.l1(wd), activation="relu"),
                Dense(units=1) # by default
    ])
    return model
```
