```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
datosNumeros=pd.read_csv("mnist_test.csv",)
```

```python
datosNumeros.shape
```

```python
tamanio=datosNumeros.shape[0]

fig,ax=plt.subplots(5,5,figsize=(10,10))

for i in range(5):
    for j in range(5):
        random=np.random.randint( tamanio )
        imageArray=np.array(datosNumeros.iloc[random,1:]).reshape((28,28))
        
        ax[i,j].set_title(str(datosNumeros.iloc[random,0] ) ) 
        ax[i,j].imshow(imageArray,aspect='auto')
        
```

# Exercise 2

* Sklearn digits dataset (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html) contains 1797 images of digits in grayscale 
* download the dataset (command is below) and:
* make an array of plots of 25 randomly chosen images (5 images x 5 images) 
* each plot has to have a title corresponding to a digit 

```python
from sklearn import datasets
digits = datasets.load_digits() # digits is a dictionary
```

```python
digits
```

```python
tamanio=digits['data'].shape[0]

fig,ax=plt.subplots(5,5,figsize=(10,10))

for i in range(5):
    for j in range(5):
        random=np.random.randint( tamanio )
        image=digits['data'][random].reshape((8,8))
        
        ax[i,j].set_title(str( digits.target[random] ) ) 
        ax[i,j].imshow(image,aspect='auto')
```

```python
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
```

```python
X_data.shape
```

```python
 y_data.shape
```

```python
dictionaryLabels={0:'airplane',
1:'automobile',
2:'bird',
3:'cat',
4:'deer',
5:'dog',
6:'frog',
7:'horse',
8:'ship',
9:'truck'}
```

```python
tamanio=X_data.shape[0]
fig,ax=plt.subplots(5,5,figsize=(10,10))

for i in range(5):
    for j in range(5):
        random=np.random.randint( tamanio )
        image=np.average(X_data,-1)[random]
        labels=dictionaryLabels[ y_data[random][0] ]
        ax[i,j].set_title( labels ) 
        ax[i,j].imshow(image,cmap='gray')
```
