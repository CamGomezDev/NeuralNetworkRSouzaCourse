### 6. Eliminando valores atípicos

Rango intercuantil

```python
df.head()
df['bedrooms'].unique(),  df['floors'].unique(),  df['waterfront'].unique(),  df['view'].unique(), df['bathrooms'].unique()
```

En las columnas bedrooms, floors, 'waterfront, view, bathrooms se realizará la eliminación de valores atípicos a mano puesto que si  se toman en cuenta, se eliminará más de la mitad de datos

```python
df_temp=df.drop(['date', 'floors', 'waterfront', 'bathrooms'], axis=1)
```

```python
Q1=df_temp.quantile(0.25)
Q3=df_temp.quantile(0.75)
IQR=Q3-Q1

IQR
```

Límite superior

```python
Q3+IQR*1.5
```

```python
Q1-1.5*IQR
```

```python
condition= ~( (df_temp> (Q3+1.5*IQR) ) | (df_temp <(Q1-1.5*IQR)  ) ).any(axis=1)
dfClean=df[condition].reset_index(drop=True)
```

```python
dfClean.head()
```

```python
print("Al eliminar los outiliers mediante el criterio de rango IQR se tienen",dfClean.shape, "datos")
print("se ha eliminado",df.shape[0]-dfClean.shape[0], "datos" )
```

```python
dfClean=dfClean[ (dfClean['bedrooms' ]>1) & (dfClean['bedrooms' ]<5) ]
dfClean=dfClean[ (dfClean['floors' ]>0) & (dfClean['floors' ]<3) ]
dfClean=dfClean[ (dfClean['bathrooms' ]>1) & (dfClean['bathrooms' ]<4 ) ]
#df['bathrooms']>0 & df['bathrooms']<6

#df['floors']>0 & df['floors']<3
```

## Preparando datos para algoritmo de NN

```python
columnsToBeUsed=['logPrice' ,'bedrooms', 'bathrooms', 'sqft_lot', 'floors',
                 'condition', 'grade', 'sqft_above', 'yr_built', 'zipcode', 'lat',
                'long','sqft_living15', 'month', 'day']

dfML=dfClean[columnsToBeUsed]
```

```python
dfML
```

```python
X=dfML.drop(['logPrice'],axis=1)
y=dfML['logPrice']

X=np.array(X)
y=np.array(y).reshape(-1,1)
```

#### Escalando datos 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_scaler=MinMaxScaler()
y_scaler=MinMaxScaler()

X_scaled=x_scaler.fit_transform(X)
y_scaled=y_scaler.fit_transform(y)
```

### Seperando datos

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_scaled, y_scaled, test_size=0.1, random_state=1234)
```

## Entrenando el algoritmo

```python
import tensorflow as tf
tf.random.set_seed(1234)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
X_train.shape
```

```python
def get_model(input_shape):
    model=Sequential(
    [InputLayer(input_shape=(input_shape)),
     Dense(units=20, activation="relu"),
     Dense(units=1, activation="linear")
    ])
    
    return model

def compile_model(model):
    model.compile(optimizer='rmsprop', loss="mae", metrics=["mae","mape"])
```

```python
model=get_model(input_shape=X_train.shape[1])
compile_model(model)
```

```python
history=model.fit(X_train, y_train, batch_size=32, epochs=500, validation_split=0.1, verbose=1)
```

```python
def plot_metrics(history):
    plt.figure(figsize=(12,6))
    plt.plot(history.history['val_mape'], label="validation")
    #plt.plot(history.history['mape'], label="training")
    plt.ylabel('Mape')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    plt.show()
    plt.close()
    
    plt.figure(figsize=(12,6))
    plt.plot(history.history['val_mae'], label="validation")
    plt.plot(history.history['mae'], label="training")
    plt.ylabel('Mae')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
```

```python
plot_metrics(history)
plt.ylim(0.04,0.06)
plt.show()
plt.close()
```

Se observa que la red neuronal en los datos de validación no está aprendiendo luego de las 20 épocas puesto presenta un comportamiento oscilatorio, además de que extrañamente los valores de MAPE en los datos de entrenamiento son de orden de $10^5$, sin embargo, los valores de MAE de los datos de entrenamiento si tienen sentido, de tal manera que se usará la métrica MAE de ahora en adelante en la evaluación. 

```python
model.evaluate(X_test,y_test)
```

#### Variando learning rate

```python
from tensorflow.keras.optimizers import RMSprop
```

```python
def get_model_learning_rate(input_shape):
    
    model=Sequential(
    [InputLayer(input_shape=(input_shape)),
     Dense(units=20, activation="relu"),
     Dense(units=1, activation="linear")
    ])
    
    return model

def compile_model_learning_rate(model, learning_rate):
    model.compile(optimizer=RMSprop(learning_rate=learning_rate), loss="mae", metrics=["mae","mape"])
```

### Mejorando el modelo


#### 1 Probando regularizadores

```python
from tensorflow.keras import regularizers
```

```python
def get_regularized_model(input_shape,weight_decay,drop_ate):
    
    model=Sequential( [
        InputLayer(input_shape=input_shape),
        Dense(units=20, activation="relu", kernel_regularizer=regularizers.l2(weight_decay) ),
        Dense(units=1, activation='linear')
    ])
    return model

def compile_model(model):
    model.compile(optimizer=RMSprop(learning_rate=0.0001), loss="mae", metrics=["mae","mape"])
```

```python
epoch=500

alpha=0.01
model1=get_regularized_model(input_shape=X_train.shape[1],weight_decay=alpha,drop_ate=0)
compile_model(model1)
history=model1.fit(X_train, y_train, batch_size=32, epochs=epoch, validation_split=0.1, verbose=0)
plt.title(f"alpha= {alpha}")
plot_metrics(history)
plt.ylim(0.04,0.06)
plt.show()
plt.close()
loss,mae,mape=model1.evaluate(X_test,y_test)
print(f"alpha:{alpha} in test data: mae={mae}, mape={mape}\n" )
#_________________________________


alpha=0.005
model1=get_regularized_model(input_shape=X_train.shape[1],weight_decay=alpha,drop_ate=0)
compile_model(model1)
history=model1.fit(X_train, y_train, batch_size=32, epochs=epoch, validation_split=0.1, verbose=0)
plt.title(f"alpha= {alpha}")
plot_metrics(history)
plt.ylim(0.04,0.06)
plt.show()
plt.close()
loss,mae,mape=model1.evaluate(X_test,y_test)
print(f"alpha:{alpha} in test data: mae={mae}, mape={mape}\n" )
#_________________________________


alpha=0.001
model1=get_regularized_model(input_shape=X_train.shape[1],weight_decay=alpha,drop_ate=0)
compile_model(model1)
history=model1.fit(X_train, y_train, batch_size=32, epochs=epoch, validation_split=0.1, verbose=0)
plt.title(f"alpha= {alpha}")
plot_metrics(history)
plt.ylim(0.04,0.06)
plt.show()
plt.close()
loss,mae,mape=model1.evaluate(X_test,y_test)
print(f"alpha:{alpha} in test data: mae={mae}, mape={mape}\n" )
#_________________________________


alpha=0.0005
model1=get_regularized_model(input_shape=X_train.shape[1],weight_decay=alpha,drop_ate=0)
compile_model(model1)
history=model1.fit(X_train, y_train, batch_size=32, epochs=epoch, validation_split=0.1, verbose=0)
plt.title(f"alpha= {alpha}")
plot_metrics(history)
plt.ylim(0.04,0.06)
plt.show()
plt.close()
loss,mae,mape=model1.evaluate(X_test,y_test)
print(f"alpha:{alpha} in test data: mae={mae}, mape={mape}" )
```

Se observa que incluyendo regularizador L2 las métricas mejoran en determinados $\alpha$. El que mejor se desempeña en los datos de testeo es: $\alpha=0.001$ 

### Implementando dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
def get_regularized_model(input_shape,weight_decay,drop_ate):
    
    model=Sequential( [
        InputLayer(input_shape=input_shape),
        Dropout(drop_ate),
        Dense(units=20, activation="relu", kernel_regularizer=regularizers.l2(weight_decay) ),
        Dropout(drop_ate),
        Dense(units=1, activation="linear")
    ])
    return model
```
