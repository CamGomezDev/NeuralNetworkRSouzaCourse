```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cargo archivos
img = pd.read_csv('mnist_test.csv') 
# genero el DataFrame
DF_img = pd.DataFrame(img) 
# columna 'label' la comvierto en array 
digitos = DF_img["label"].to_numpy() 
 # elimino la columna de 'label'
dd = DF_img.drop(columns=['label']) 
# elimino la primera fila
dd = dd.drop([0,0])   
dd   # imprimo el DataFrame
```

```python
type(digitos)
```

```python
fig, ax = plt.subplots(5,5,figsize=(8,8))   # Genero las graficas
azar = np.random.randint(0,9999,25).reshape(5,5) # matriz (5x5) de números enteros aleatorios.

for i in range(0,5):
    for j in range(0,5):
        n = azar[i,j]  
        graf = dd.iloc[n].to_numpy()
        graf = graf.reshape(28,28)  
        ax[i,j].set_title(digitos[n+1])
        ax[i,j].imshow(graf)
        ax[i,j].set_axis_off()
        
        
fig.tight_layout(pad=1.0) 
```
