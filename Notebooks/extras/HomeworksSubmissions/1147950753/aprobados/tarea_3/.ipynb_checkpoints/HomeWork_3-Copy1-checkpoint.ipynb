* Realize numerically gradient descent method on the "data.csv" which represent some linear regression problem (fitting) Use following parameters:


$\theta_{1} = 10$  #initial values of   
$\theta_{0} = 10 $    # y=theta_1*x + theta_0  
$\alpha = 0.0001 $  # The learning Rate  
epochs = 100     # The number of iterations  


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
from scipy.optimize  import curve_fit

data = pd.read_csv('data.csv',header = None)
#DF_data = pd.DataFrame(data, columns= ['X','Y'])

x = data[0]
y = data[1]

# ploteo los datos

fig, ax = plt.subplots()
ax.plot(x,y,'+')
```

```python
#  Gradiente descendente

def Sumatoria_x0(xx,yy,t_0,t_1,m):
    s = 0
    for i in range(0,len(xx)):
        s = s + t_1*x[i] + t_0 - y[i]
        #s += s 
    return s

    
def Sumatoria_x1(xx,yy,t_0,t_1,m):
    s = 0
    for i in range(0,len(xx)):
        s = s + (t_1*x[i] + t_0 - y[i])*x[i]
        #s += s 
    return s

```

```python
def fuu(tetha_0,tetha_1, alfa ):
    
    #alfa = 0.0001 
    tetha_0 = 10
    tetha_1 = 10
    m=100
    
    MAE = []
    MSE = []
    MAPE = []

    for n in range(0,100):
        
        
        tetha_0 = tetha_0 - alfa*Sumatoria_x0(x,y,tetha_0,tetha_1,len(x))/m
        tetha_1 = tetha_1 - alfa*Sumatoria_x1(x,y,tetha_0,tetha_1,len(y))/m
        
        MSE.append(sum(tetha_1*x + tetha_0 - y))
        MAE.append(abs(sum(tetha_1*x + tetha_0 - y)))
        MAPE.append(abs(sum((tetha_1*x + tetha_0 - y) /x)))
             
    return tetha_1,tetha_0 , MSE, MAE ,MAPE

# Llamo la funcion fuu
tetha_1, tetha_0, MSE, MAE ,MAPE = fuu(10,10,alfa = 0.0001)
 
print("tethe_0 = %.2f / tetha_1 = %.2f"%(tetha_0,tetha_1))
```

### After finding the optimal and that minimized the MSE function make four plots (on an array... using plt.subplots(2,2))

* plot 1 the final fitting line to the data and compare it with the scipy.optimize (lm) build-in function
* plot 2 time evolution of MSE vs number of iterations
* plot 3 time evolution of MAE vs number of iterations
* plot 4 time evolution of MAPE vs number of iterations


```python
# plot 1 the final fitting line to the data and compare it with the scipy.optimize (lm) build-in function
fig, ax = plt.subplots(1,2,figsize=(10,6)) 


xx = np.linspace(0,70,len(x)+33)

tetha_0 = 9.83
tetha_1 = 1.29


def final_fitting(xx):
    yy = []
    for i in xx:
        y = tetha_1*i + tetha_0
        yy.append(y)
    return yy

 

ax[0].plot(xx,final_fitting(xx))
ax[0].set_title("Gradiente descendente")
ax[0].plot(x,y,".")



###################################
###Scipy.optimize
#################################



def cuadratic(x,a,b):
    
    return  a*x + b

[aa,bb] , cov = curve_fit(cuadratic,x,y)

#print()

ax[1].plot(x,cuadratic(x,aa,bb),label='Curve-fit') 
ax[1].set_title("Curve-fit")
ax[1].plot(x,y,".")
```

```python
print("Obtenidos mediante Curve-fit tetha_1= %.2f / tetha_0 = %.2f"%(aa,bb))
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,6))

ax[0].plot(MAE) 
ax[0].set_title("MAE vs number of iterations")
     
ax[1].plot(MSE)
ax[1].set_title("MSE vs number of iterations")

ax[2].plot(MAPE)
ax[2].set_title("MAPE vs number of iterations")

fig.tight_layout()
plt.show()
```
