# Exercise 1
* the file 'mnist_test.csv' contains 10000 images of digits in grayscale saved as a rows in the csv file
* first column contains value of the digits then 784 columns correspond to pixels
* make an array of plots of 25 randomly chosen images (5x5) 
* each plot has to have a title corresponding to a digit 

```python
!ls
```

```python
im_values = pd.read_csv('mnist_test.csv')
```

```python
fig, ax = plt.subplots(5,5,figsize=(18,18))

for i in range(5):
  for j in range(5):
    rand = np.random.randint(0, 10000)
    data_image = im_values.loc[rand].to_numpy() # to extract pixel values of image in random row
    ax[i,j].set_title(data_image[0])
    data_image = np.reshape(data_image[1:785], (28,28)) # to reshape in a 2D array
    ax[i,j].imshow(data_image, cmap='gray', vmin=0, vmax=255)
```

* make an array of plots of 25 randomly chosen images (5 images x 5 images) 
* each plot has to have a title corresponding to a digit 

```python
digits = datasets.load_digits() # digits is a dictionary
print(digits['data'])
```

```python
fig, ax = plt.subplots(5,5,figsize=(18,18))

for i in range(5):
  for j in range(5):
    rand = np.random.randint(0, 1797)
    data_image = digits['data'][rand] # to extract pixel values of image in random row of key 'data'
    data_image = np.reshape(data_image, (8,8)) # to reshape in a 2D array
    ax[i,j].set_title(digits['target'][rand])
    ax[i,j].imshow(data_image, cmap='gray', vmin=0, vmax=255)
```

* make an array of plots of 25 randomly chosen images (5 images x 5 images) 
* each plot has to have a title with a corresponding label

```python
(x_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data() # We use only the training sets
```

```python
classes = {0:'Airplane', 1:'Automobile', 2:'Bird', 3:'Cat', 4:'Deer', 5:'Dog', 6:'Frog', 7:'Horse', 8:'Ship', 9:'Truck'}
```

```python
fig, ax = plt.subplots(5,5,figsize=(18,18))

for i in range(5):
  for j in range(5):
    rand = np.random.randint(0, 50000)
    data_image = x_data[rand] # to extract pixel values of random image
    ax[i,j].set_title(str(y_data[rand][0]) + '. ' + classes[y_data[rand][0]])
    ax[i,j].imshow(data_image)
```

```python
# Using axis option of Numpy mean function, one can get the mean value of each pixel
# instead of get the mean value of all the array or the mean of each row
x_data_gray = np.mean(x_data, axis=3)
```

```python
fig, ax = plt.subplots(5,5,figsize=(18,18))

for i in range(5):
  for j in range(5):
    rand = np.random.randint(0, 50000)
    data_image = x_data_gray[rand]
    ax[i,j].set_title(str(y_data[rand][0]) + '. ' + classes[y_data[rand][0]])
    ax[i,j].imshow(data_image, cmap='gray', vmin=0, vmax=255)
```
