## Solution E1


```python
#Import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read image file using pandas

mnist_test = pd.read_csv("mnist_test.csv")
#mnist_test
```

```python

#Convert pandas dataframe to numpy array

X_data = mnist_test.loc[:,'1x1':'28x28'].to_numpy().reshape((10000, 28, 28))
y_data = mnist_test.loc[:,'label'].to_numpy()

#Choose n_rows and n_columns as you prefer
n_rows = 5
n_columns = 5

#Random selection of a sample of size 25
random_sample = np.random.randint(0, X_data.shape[0], size=(n_rows*n_columns)).reshape(n_rows,n_columns)

#Plotting digits in a figure n_rows x n_columns

fig, ax = plt.subplots(n_rows, n_columns, figsize=(8,8))

for i in range(n_rows):
    for j in range(n_columns):
        
        ax[i][j].set_title("Digit " + str(y_data[random_sample[i][j]]))
        ax[i][j].imshow(X_data[random_sample[i][j]], cmap="gray")
        ax[i][j].axis('off')

fig.tight_layout(pad=1.0)
```

```python
#Charging library of datasets dictionaries
from sklearn import datasets
#Choose 'load_digits'
digits = datasets.load_digits() # digits is a dictionary
```

```python
#Convert dataset dictionary to numpy array

X_data = digits['data'].reshape((digits['data'].shape[0], 8, 8))
y_data = digits['target']

#Choose n_rows and n_columns as you prefer
n_rows = 5
n_columns = 5

#Random selection of a sample of size 25
random_sample = np.random.randint(0, X_data.shape[0], size=(n_rows*n_columns)).reshape(n_rows,n_columns)

#Plotting digits in a figure n_rows x n_columns

fig, ax = plt.subplots(n_rows, n_columns, figsize=(8,8))

for i in range(n_rows):
    for j in range(n_columns):
        
        ax[i][j].set_title("Digit " + str(y_data[random_sample[i][j]]))
        ax[i][j].imshow(X_data[random_sample[i][j]], cmap="gray")
        ax[i][j].axis('off')

fig.tight_layout(pad=1.0)
```

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
```

```python
X_data.shape
```

```python
labels = {
    0: "airplane",
    1: "automobile",
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck"
}
```

```python
#Choose n_rows and n_columns as you prefer
n_rows = 5
n_columns = 5

#Random selection of a sample of size 25
random_sample = np.random.randint(0, X_data.shape[0], size=(n_rows*n_columns)).reshape(n_rows,n_columns)

#Plotting digits in a figure n_rows x n_columns

fig, ax = plt.subplots(n_rows, n_columns, figsize=(8,8))

for i in range(n_rows):
    for j in range(n_columns):
        
        ax[i][j].set_title("Name: " + labels[int(y_data[random_sample[i][j]])])
        ax[i][j].imshow(X_data[random_sample[i][j]])
        ax[i][j].axis('off')

fig.tight_layout(pad=1.0)
```
