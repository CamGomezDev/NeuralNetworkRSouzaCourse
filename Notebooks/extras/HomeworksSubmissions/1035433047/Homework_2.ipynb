##Exercise 1

```python
df=pd.read_csv('mnist_test.csv',sep=',')
df_np=df.to_numpy()
print(df_np.shape)
```

```python
X_data=df_np[:,1:]
y_data=df_np[:,0]
```

```python
print(X_data.shape)
print(y_data.shape)
```

```python
ran_idx=np.random.randint(0,df_np.shape[0],25)
ran_idx
```

```python
X_data_selected=X_data[ran_idx]
y_data_labels=y_data[ran_idx]
```

```python
print(X_data_selected.shape)
print(y_data_labels.shape)
```

```python
lista=[]
for i in range(len(X_data_selected)):
  x=np.reshape(X_data_selected[i],(28,28))
  lista.append(x)
lista=np.array(lista)
print(lista.shape)
```

```python
fig, ax = plt.subplots(5,5,figsize=(17,17))

for i in range(5):
  for j in range(5):
    img=lista[5*i+j]
    ax[i,j].set_title('NÃºmero'+' '+str(y_data_labels[5*i+j]))
    plt.gray()
    ax[i,j].imshow(img)
```

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
```

```python
print(X_data.shape)
print(y_data.shape)
```

```python
ran_idx2=np.random.randint(0,X_data.shape[0],25)
ran_idx2
```

```python
X_data2_selected=X_data[ran_idx2]
y_data2_labels=y_data[ran_idx2]
```

```python
fig, ax = plt.subplots(5,5,figsize=(16,16))

for i in range(5):
  for j in range(5):
    img=X_data2_selected[5*i+j]
    img_avg=np.average(img,-1)
    ax[i,j].set_title('NÃºmero'+' '+str(y_data2_labels[5*i+j][0]))
    ax[i,j].imshow(img_avg)
```
