```python
def c2gray1(M):
    """
    This function converts a (32,32,3) colored image into a (32x32) gray image.
    
    Parameters
    ----
    
    M : Array-like
        32x32x3-shape array to convert.
        
    Return
    ----
    
    c : Array-like
        The new array of shape 32x32 gray-scale.
    """
    b=M.copy().transpose(2,0,1)
    gray=(1/(3*255))*(b[0]+b[1]+b[2])
    return gray
```

```python
#Constructing a new set of images converting the colored images into gray images.
nX_data3=[]
for i in range(0,X_data3.shape[0]):
    nX_data3.append(c2gray1(X_data3[i]))
```

```python
fig3_gray,ax3_gray=plt.subplots(5,5,figsize=(6,6))
for i in range(0,ran_indx_matrix3.shape[0]):
    for j in range(0,ran_indx_matrix3.shape[1]):
        #the ij entry of this matrix is the random index of the image.
        k=ran_indx_matrix3[i,j]
        labl=labels[y_data3[k][0]]
        #ploting the image.
        ax3_gray[i,j].imshow(nX_data3[k],cmap="gray")
        ax3_gray[i,j].set_title(labl)
        ax3_gray[i,j].set_axis_off()
        
#setting the separation between subplots.        
fig3_gray.tight_layout(pad=1.0)
```

```python
#selection of a random sample of 25 elements from gray images.
n_samples3_gray=np.random.randint(0,len(nX_data3),25)

#converting the 1D array of samples into a 2D array of 5x5.
ran_indx_matrix3_gray=n_samples3_gray.reshape(5,5)
```
