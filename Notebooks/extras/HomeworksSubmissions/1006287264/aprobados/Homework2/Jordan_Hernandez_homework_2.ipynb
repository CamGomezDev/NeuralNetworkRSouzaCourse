# Exercise 1
* the file 'mnist_test.csv' contains 10000 images of digits in grayscale saved as a rows in the csv file
* first column contains value of the digits then 784 columns correspond to pixels
* make an array of plots of 25 randomly chosen images (5x5) 
* each plot has to have a title corresponding to a digit 

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('mnist_test.csv')
df.head()
#df.shape
```

```python
# Convirtiendo datos a formato numpy
df_np = df.to_numpy()
df_np.shape
# el array y tiene las etiquetas 
y_data = df[['label']].to_numpy()
# El array que contiene los números 
X_data=df.loc[:, df.columns.drop('label')].to_numpy()
```

```python
X_data[0].shape
```

```python
# Generando números aleatorios 
num_samples = X_data.shape[0]
#print(num_samples)
rand_idx = np.random.randint(0,num_samples,25)
rand_idx
```

```python
X_random= X_data[rand_idx]
y_random= y_data[rand_idx]
```

```python
#Graficas 
_, axes = plt.subplots(nrows=5,ncols=5, figsize=(10, 8))
for ax, image, label in zip(axes.ravel(),X_random, y_random):
    ax.set_axis_off()
    image = image.reshape(28, 28)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="none" )
    ax.set_title("Número: %i" % label)
```

```python
#cambiando de formato unit a int 
y_names= y_data.astype(int).ravel()
type(y_names.ravel()[0])
```

```python
#Dandole nombre a las posibles clases
class_names = ["airplane", " automobile", "bird", "cat", "deer",
               "dog", "frog", "horse", "ship", "truck"]
class_names[y_names[0]]
```

```python
idx=np.random.randint(0,X_data.shape[0],25)
idx
```

```python
# Tomando los arrays aleatorios 
X_random= X_data[idx]
y_random= y_names[idx]
```

```python
_, axes = plt.subplots(nrows=5,ncols=5, figsize=(10, 8))
for ax, image, label in zip(axes.ravel(),X_random, y_random):
    ax.set_axis_off()
    #image = image.reshape(32, 32,3)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="none" )
    ax.set_title("Objeto: %s" % class_names[label])
```

```python
# Convirtiendo a escala de grises 
```

```python
X_avg= np.average(X_data,-1)
index=np.random.randint(0,X_avg.shape[0],25)
index

X_avg_random= X_avg[index]
y_avg_random= y_names[index]
```

```python

_, axes = plt.subplots(nrows=5,ncols=5, figsize=(10,8) )
for ax, image, label in zip(axes.ravel(),X_avg_random, y_avg_random):
    ax.set_axis_off()
    image = image.reshape(32, 32,1)
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="none" )
    ax.set_title("%s" % class_names[label])
```
