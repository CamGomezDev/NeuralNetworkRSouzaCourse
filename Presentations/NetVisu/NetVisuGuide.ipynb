```python
import NetVisu
import gudhi
import matplotlib.pyplot as plt
```

# 2 entangled circles

```python
data, target = NetVisu.datasets.entangled_two_circles(200, True)

model_relu = NetVisu.feedforward(4, 15, 'relu')
NetVisu.trains_network(model_relu, 100, data, target, verbose = 0, plot = True)
```

```python
model_relu.evaluate(data, target)
```

```python
NetVisu.pca_network(model_relu, data, target)
```

```python
NetVisu.persistence_homology(data)
```

```python
NetVisu.persis_homol_network(model_relu, data, target)
```
